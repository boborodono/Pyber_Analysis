```python
# Import dependencies
import matplotlib
import matplotlib.pyplot as plt
matplotlib.__version__
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis, color="green", marker="D", linewidth=2)
# Change the line color to green.
# Change the marker to a diamond.
# Change the line width to 2 points.
# Add a legend for the city of Boston.
ax.legend("Boston")
# Add a title and axes labels.
ax.set_title("Pyber Fare by Month")
ax.set_ylabel("Fare($)")
ax.set_xlabel("Month")
# Add a y-axis limit.
ax.set_ylim(0,45)
# Add grid lines.
ax.grid()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="+", color="blue", linewidth=2, label ='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("Pyber Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis,
           x_axis,
           s = [i * 5 for i in y_axis],
           c="skyblue",
           linewidth=2,
           label = "Boston",
           alpha=0.8,
           edgecolors="black",)
ax.legend()
plt.title("PyBer Fare by Month")
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.xlim(0,50)
plt.gca().invert_yaxis()
```

```python
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```
